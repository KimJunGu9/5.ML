# 1. Iris 데이터 셋

* https://scikit-learn.org/stable/modules/classes.html?highlight=datasets#module-sklearn.datasets

```python
from sklearn.datasets import load_iris
import pandas as pd
```

```python
iris = load_iris()
```

```python
iris
```

```python
print(iris['DESCR'])  # 설명을 따로 불러와서 찍어보기
```

```python
# sepal length in cm : 꽃받침의 길이
# sepal width in cm : 꽃받침의 너비
# petal length in cm : 꽃잎의 길이
# petal width in cm : 꽃잎의 너비

data = iris['data']
data[:5]
```

```python
feature_names = iris['feature_names']
feature_names  # 필드네임을 가져오기
```

```python
df_iris = pd.DataFrame(data, columns=feature_names) # data를 넣고 columns명에 feature_name을 넣는다.
# 1차원 데이터를 2차원 표로 만드는 방법은 pd.DataFrame()에다가 데이터를 넣고, 컬럼 이름을 넣는다.
```

```python
df_iris.head()
```

```python
target = iris['target']  # target : 정답
target[:5]
```

1.png

```python
df_iris['target'] = target # 파생변수 만들기 # 열 추가 (columns 값 추가)
df_iris.head()
```

2.png

```python
df_iris
```

```python
from sklearn.model_selection import train_test_split  # train_test_split : 학습데이터와 테스트데이터를 분류해주는 역할

# sklearn에 model_selection이라는 클래스가 있고 거기에 함수중에 train_test_split이라는 함수가 있다.
```

```python
x_train, x_test, y_train, y_test = train_test_split(  
    df_iris.drop('target', 1), df_iris['target']   
)

# x_train : 학습을 시킬 데이터
# x_test : 학습에 필요한 정답
# y_train : 테스트를 시킬 데이터
# y_test : 테스트에 필요한 정답

# df_iris.drop('target', 1) :  타겟을 제외하고 데이터를 넣는 것, df_iris['target'] : 타겟만 넣어서 데이터를 확인

# drop() : 필드를 삭제하는 것
```

```python
x_train.shape, y_train.shape # 112개, 4개의 필드, 112개, 1개의 필드

# x_train.shape : 학습시키기 위한 데이터가 몰려있음 (데이터만 들어가 있음)
# y_train.shape : 정답이 들어가 있음
```

```python
x_test.shape, y_test.shape # 38개, 4개의 필드, 38개, 1개의 필드

# x_valid.shape : 검증용 데이터
# y_valid.shape : 정답이 들어가 있음
```

```python
x_train, x_test, y_train, y_test = train_test_split( 
    df_iris.drop('target', 1), df_iris['target'], stratify=df_iris['target']  # stratify : 균등하게 배분하고싶으면
)
```

```python
sns.countplot(y_train)
```

```python
x_train.shape, y_train.shape
```

```python
x_test.shape, y_test.shape
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
svc = SVC()
svc.fit(x_train, y_train)
```

```python
y_pred = svc.predict(x_test)
print('정답률 : ', accuracy_score(y_test, y_pred))
```
